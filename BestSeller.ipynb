```python
data=pd.read_csv('bestsellers.csv')
data.head()
```

```python
# Retrieve a series that contain a book name. 
data['Name'].values  # same with data['Name]
```

```python
# Retrieve a series that contains user ratings. 


data['User Rating'].values
```

```python
# Retrieve the first 8 author. 


data.loc[0:7,'Author'] # loc way
data['Author'].head(n=8) # head way
```

```python
# Find the unique genres



data['Genre'].unique()
```

```python
# Find the no. Of unique authors


data['Author'].nunique()
```

```python
# Find the avg. Price


data['Price'].mean()
```

```python
# Find the 10 highest price

data.nlargest(10,'Price') # can also be done by data.sort_values('Price) then new.head(n=10)
```

```python
# Find the three most common titles in the dataset.. 
data['Name'].value_counts().nlargest(n=3)
```

```python
# Create a new dataframe with only author or user rating
new =data[['Author','User Rating']]
new
```

```python
# Using the new dataframe find the most common combination of author and user rating score.. 


new.groupby(['Author','User Rating'])['User Rating'].value_counts().nlargest(n=1)
```

```python
# Find the top 10 most prolific author and plot their no of books as a bar chart.

import matplotlib.pyplot as plt

df=data.groupby('Author').size()

df.plot.bar(list(df.keys()))


```
